```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
df
```

```python
input = df.drop(['Survived'], axis = 'columns')
```

```python
input
```

```python
target = df['Survived']
target
```

```python
dummies = pd.get_dummies(input.Sex).astype(int)
dummies
```

```python
merged = pd.concat([input, dummies], axis = 'columns')
merged
```

```python
merged.drop(['Sex'], axis = 'columns', inplace = True)
merged
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(merged, target)
```

```python
model.score(merged, target)
```

```python
model.predict([[1, 38.0, 71.2833, 1, 0]])
```
