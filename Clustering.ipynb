```python
# Clustering: Unsupervised Machine Learning where the GOAL is to determine hidden clusters(groups) within the dataset.
# Step1: Read Data
import pandas
data = pandas.read_csv("https://modcom.co.ke/data/datasets/AirlinesCluster.csv")
data
```

```python
# Grouping Data Into Different Clusters based on Similarities
# E,g Safaricom by grouping their customers based on Calls, sms, bundles
array = data.values
dataValues = array[:, 4:7]
data = pandas.DataFrame(dataValues, columns = ["FlightMiles",	"FlightTrans",	"DaysSinceEnroll"])
data
```

```python
# Split Data Into X
array = data.values
X = array[:, :]
X
```

```python
# Import KMeans(Group data based on the mean)
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5, random_state=42)
clusters = model.fit(X)
clusters
```

```python
# Mean Clusters(Centronoids)
means = model.cluster_centers_
groups = pandas.DataFrame(means, columns = ["FlightMiles",	"FlightTrans", 	"DaysSinceEnroll"])
groups
```

```python
data['Cluster'] = model.labels_
```

```python
data
```

```python
# Getting Members of the Same Clusters
# E,g Cluster 4 Members
cluster_4_members = data[data['Cluster'] == 4]
cluster_4_members
```
