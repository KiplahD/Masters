```python
# Simple Deep Model:
# Regression

import pandas
data = pandas.read_csv("Advertising.csv")
data
```

```python
# Prediction - Supervised Learning
# Divide X and Y
array = data.values

X = array[:, 0:3]
Y = array[:, 3]
```

```python
# Split Data into Training Set and Testing Set
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# How is the Learning?
# ML - applied scientific algoriths e,g DecisionTree, KNeighbors....
# DL - Construct an Artificial Neural Network - keras
```

```python
# ANN Features: Sequential, Dense, Dropout, Flatten, activation function
from keras.models import Sequential
from keras.layers import Dense, Dropout
import keras
```

```python
# Used to generate sequence of nodes(neurons)
model = Sequential()
```

```python

# Create the Layers using Dense

# Read more: relu, softmax, sigmoid activation functions
model.add(Dense(500, input_dim = 3, activation = "relu"))

model.add(Dense(100, activation = "relu"))

model.add(Dense(50, activation = "relu"))


# Output Layer
model.add(Dense(1, activation = "sigmoid"))
```
