```python
#Basket Analysis: Divided in to Supervised and Unsupervised machine learning.
#Supervised machine learning - regression, classification
#Unsupervised mahine learning - Clustering, association rules/data mining
#Baasket analysis: is a technique that applies association rule(apriori) to discover relationship between items
```

```python
#Data collection
import pandas
data = pandas.read_csv("https://modcom.co.ke/data/datasets/store.csv")
data
```

```python
#Import association algorith (apriori)
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Data analysis
#1. Transactions per country
data.groupby('Country').size()
```

```python
#Market Analysis: Purchasing depends on location
#For instance you are in urban setting: pizza, canned goods.
#While on rural setting: vegetables

#Clothes that are bought in the US are very different from clothes that are bought in Arabian Countries
#Adjust a particular location

data = data[ data['Country'] == 'France']
data
```

```python
#Null records
data.isnull().sum()
#66 customers in France transaction paid by cash(ID Not recorded)
```

```python
#Groupby InvoiceNo, Description alongside total Quantity
#A negative quantity represents the return
data.groupby(['InvoiceNo', 'Description']) ['Quantity'].sum()
```

```python
#Create a pivot table and dump all the transactions(single pool)
#index - InvoiceNo
#columns - description
#values - quantity
#Replace Quantity Nan with 0

pandas.set_option('display.max_columns', 300)
baskets = data.pivot_table(index= 'InvoiceNo', columns ='Description', values = 'Quantity', fill_value=0)
baskets
```

```python
#Any item that has a quantity that is more than 1: we will mark that it was bought (label it to 1)
#0 represent item not bought and 1 represents an item was bought.

#create a function(if quantity > 1 mark1 else mark 0)
def encode(x):
  if x <= 0:
    return 0
  if x > 0 :
    return 1
```

```python
#Apply  to the dataset pivot
basket_set = baskets.applymap(encode)
basket_set
```

```python
#Apply the association rule(apriori)
#frequent items
frequent_items_sets = apriori(basket_set, min_support=0.10, use_colnames=True)
rules = association_rules(frequent_items_sets, metric = 'lift')
rules[(rules['lift']>7) & (rules['confidence'] > 0.07)]
```
