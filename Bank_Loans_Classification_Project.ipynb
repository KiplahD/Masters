```python
import pandas
data = pandas.read_csv("https://modcom.co.ke/data/datasets/loans.csv")
data
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['credit.policy'] = encoder.fit_transform(data['credit.policy'])
data['purpose'] = encoder.fit_transform(data['purpose'])
data['int.rate'] = encoder.fit_transform(data['int.rate'])
data['installment'] = encoder.fit_transform(data['installment'])
data['log.annual.inc'] = encoder.fit_transform(data['log.annual.inc'])
data['dti'] = encoder.fit_transform(data['dti'])
data['fico'] = encoder.fit_transform(data['fico'])
data['revol.bal'] = encoder.fit_transform(data['revol.bal'])
data['revol.util'] = encoder.fit_transform(data['revol.util'])
data['inq.last.6mths'] = encoder.fit_transform(data['inq.last.6mths'])
data['delinq.2yrs'] = encoder.fit_transform(data['delinq.2yrs'])
data['pub.rec'] = encoder.fit_transform(data['pub.rec'])
pandas.set_option('display.max_columns', None)
data
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
estimator = DecisionTreeClassifier()

num_of_features = 10
rfe = RFE(estimator=estimator, n_features_to_select=num_of_features)

array = data.values
A = array[:, 0:12] # X
B = array[:, 12] # Y

rfe.fit(A, B)
print("Selected Features ", rfe.support_)
```

```python
subset = data[['credit.policy', 'int.rate', 'installment', 'log.annual.inc', 'dti', 'fico', 'revol.bal', 'revol.util', 'inq.last.6mths', 'delinq.2yrs', 'not.fully.paid']]
subset
```

```python
array = subset.values
X = array[:, 0:10]
Y = array[:, 10]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaledX = scaler.fit_transform(X)

from imblearn.over_sampling import SMOTE
oversample =  SMOTE()
x, y = oversample.fit_resample(X, Y)
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB

models = []
models.append(('KNN', KNeighborsClassifier()))
models.append(('DTC', DecisionTreeClassifier()))
models.append(('GBC', GradientBoostingClassifier()))
models.append(('RFC', RandomForestClassifier()))
models.append(('SVC', SVC()))
models.append(('LR', LogisticRegression(max_iter=1000)))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('NB', GaussianNB()))


for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=42, shuffle=True)
    cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv = kfold, scoring="accuracy")
    print(name, cv_results.mean())
```

```python
model = RandomForestClassifier()
model.fit(X_train, Y_train)
print('Training.....')

predictions = model.predict(X_test)
print("Model Predictions", predictions)
print("Y Test ", Y_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(accuracy_score(predictions, Y_test))
print(classification_report(predictions, Y_test))
print(confusion_matrix(predictions, Y_test))
print(data.groupby('not.fully.paid').size())
```
