```python
#Web scraping is technique in data science of extracting information from websites available on the internet
#Digital economy: most of organizations, businesses, journals research papers are readily available on the internet
#As a data scientist we need a way to extract these valuable from the internet, perform analysis and get the insights
#Requirement: websites are made of tags(<openingtag>)

#Step 1
#Enter the link and the Parser element
from bs4 import BeautifulSoup
import requests

URL = "https://www.bbc.com/news/live/world-middle-east-67258466"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
paragraphs = soup.find_all("p") # Find all paragraph tags
text = ""
for p in paragraphs:
    text += p.get_text() + "\n" # Concatenate text from each paragraph

print(text)
```

```python
import nltk
nltk.download('punkt')
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize, word_tokenize
tokenized_text = word_tokenize(text)
print (tokenized_text)
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import  stopwords
print(stopwords.fileids())

english_stopwords = set(stopwords.words('english'))
print(english_stopwords)
```

```python
meaningful_words = []
for word in tokenized_text:
  if word.lower() not in english_stopwords:
    meaningful_words.append(word)
print(meaningful_words)
```

```python
clean_words = []
for word in meaningful_words:
  if word.isalpha():
    clean_words.append(word)
print(clean_words)
```

```python
from nltk.probability import FreqDist
frequency_count= FreqDist(clean_words)
frequency_count.most_common(20)
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image

#generate  a function  to generate WordCloud
def generate_wordcloud(text):
  stopwords = set(STOPWORDS)
  wc = WordCloud(background_color= "white", max_words=1000, stopwords=stopwords, repeat=True)
  wc.generate(str(text))
  wc.to_file("wc.png")
  path = "wc.png"
  display(Image.open(path))
generate_wordcloud(frequency_count.most_common(20))
```
